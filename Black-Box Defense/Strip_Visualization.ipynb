```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

def remove_strip(image, orientation='horizontal', strip_size=10):
    if orientation == 'horizontal':
        # Choose a random horizontal strip to remove
        y = np.random.randint(0, image.shape[0] - strip_size)
        image[y:y+strip_size, :, :] = 0
    else:
        # Choose a random vertical strip to remove
        x = np.random.randint(0, image.shape[1] - strip_size)
        image[:, x:x+strip_size, :] = 0
    return image

# Load an image
image_path = 'ocular_recognition/test/1141/1141_l_1.png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply strip removal
orientation = 'horizontal' if np.random.rand() < 0.5 else 'vertical'
augmented_image = remove_strip(image.copy(), orientation="horizontal")

# Display the original and augmented images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[1].imshow(augmented_image)
ax[1].set_title('Image with Strip Removal')
plt.show()
```
